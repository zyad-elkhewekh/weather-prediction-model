```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_pickle('data.pkl')
```

```python
x_train, x_test, y_train, y_test = train_test_split( data, data, test_size = 0.3, random_state=42)




model = LinearRegression()
model.fit(x_train, y_train)

# importance = pd.Series(model.coef_, index=x_train.columns)
# importance = importance.abs().sort_values(ascending=False)
# print(importance)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn import metrics

print("MSE:", metrics.mean_squared_error(y_test, y_pred))
print("MAE:", metrics.mean_absolute_error(y_test, y_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("RÂ² Score:", metrics.r2_score(y_test, y_pred))
```

```python
from sklearn.metrics import*
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
